```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
import seaborn as sns
```

```python
df = pd.read_csv('fcc-forum-pageviews.csv',index_col='date', parse_dates=True)
df.head()
```

```python
df = df[(df['value'] >= df['value'].quantile(0.025)) & (df['value'] <= df['value'].quantile(0.975))] 
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.lineplot(df,x='date', y='value',ax=ax).set(xlabel='Date', ylabel='Page Views',title='Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
```

```python
df['month'] = df.index.month_name()
df['year'] = df.index.year
```

```python
month_order= ['January' ,'February','March','April','May','June','July','August','September','October','November','December']
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.color_palette()
sns.set_theme(style='darkgrid')
sns.barplot(df,x='year',y='value',hue='month',hue_order=month_order,errorbar=None).set(xlabel='Years',ylabel='Average Page Views')
```

```python
df_ba = df.copy()
df_ba['month'] = df_ba.index.month_name()
df_ba['year'] = df_ba.index.year
```

```python
df_ba = df_ba.groupby(['year','month'])['value'].mean().unstack()
df_ba = df_ba[month_order]
```

```python
fig = df_ba.plot(
    kind='bar',
    xlabel='Years',
    ylabel='Average Page Views',
    figsize=(10,8)
).figure

plt.legend(title='Months')
```

```python
df_box = df.copy()
df_box.reset_index(inplace=True)
df_box['year'] = [d.year for d in df_box.date]
df_box['month'] = [d.strftime('%b') for d in df_box.date]
```

```python
df_box.head()
```

```python
month_ord= ['Jan' ,'Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
fig,axe = plt.subplots(figsize=(20,7),nrows=1, ncols=2)

sns.set_theme(style='darkgrid')

sns.boxplot(df_box,x='year', y='value',ax=axe[0],hue='year',palette='bright',legend=False)
axe[0].set(xlabel='Year',ylabel='Page Views',title='Year-wise Box Plot (Trend)')



sns.boxplot(df_box,x='month', y='value',order=month_ord,ax=axe[1],hue='month',palette='bright',legend=False)
axe[1].set(xlabel='Month',ylabel='Page Views',title='Month-wise Box Plot (Seasonality)')
```

# Sea level predictor

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.head()
```

```python
fig = df.plot(
    kind='scatter',
    x='Year',
    y='CSIRO Adjusted Sea Level'
).figure
```

```python
# 1. Use linregress to get the slope and y-intercept
res = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
print(f"Slope: {res.slope}, Intercept: {res.intercept}")

# 2. Create an array of years spanning from the first year in the dataset up to 2050
years_extended = pd.Series(range(df['Year'].min(), 2051))

# 3. Calculate the predicted sea levels using the line equation: y = mx + b
predicted_sea_levels = res.slope * years_extended + res.intercept

# 4. Plot the scatter plot and the new predicted line of best fit together
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.plot(years_extended, predicted_sea_levels, color='red', label='First line of best fit')
plt.legend()
```

```python
res = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
years = pd.Series(range(df['Year'].min() , 2051))
equ = res.slope * years + res.intercept
print(f"Slope: {res.slope}, Intercept: {res.intercept}")
```

```python
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
plt.plot(years, equ , color= 'red')
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
```

```python
df_sec = df[df['Year'] >= 200]
resn = linregress(df_sec['Year'], df_sec['CSIRO Adjusted Sea Level'])
nYear = pd.Series(range(df_sec['Year'].min() , 2051))
new = resn.slope * nYear + resn.intercept
```

```python
plt.scatter(df_sec['Year'],df_sec['CSIRO Adjusted Sea Level'])
plt.plot(nYear, new , color= 'red')
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
```
